```python
from sklearn.preprocessing import LabelEncoder

df_numeric = df.copy()

# 타깃 변수 숫자로 인코딩 (HadHeartAttack: Yes/No -> 1/0)
df_numeric['HadHeartAttack'] = df_numeric['HadHeartAttack'].map({'No': 0, 'Yes': 1})

# 숫자로 매핑
ordinal_mappings = {
    'GeneralHealth': {'Poor': 0, 'Fair': 1, 'Good': 2, 'Very good': 3, 'Excellent': 4},
    'AgeCategory': {
        'Age 18 to 24': 0, 'Age 25 to 29': 1, 'Age 30 to 34': 2, 'Age 35 to 39': 3,
        'Age 40 to 44': 4, 'Age 45 to 49': 5, 'Age 50 to 54': 6, 'Age 55 to 59': 7,
        'Age 60 to 64': 8, 'Age 65 to 69': 9, 'Age 70 to 74': 10, 'Age 75 to 79': 11,
        'Age 80 or older': 12
    }
}

for col, mapping in ordinal_mappings.items():
    df_numeric[col] = df_numeric[col].map(mapping)

# LabelEncoderOrdinal 인코딩으로 명목 범주형 열 인코딩
nominal_cols = [
    'State', 'Sex', 'LastCheckupTime', 'PhysicalActivities', 'RemovedTeeth',
    'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer', 'HadCOPD',
    'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis', 'HadDiabetes',
    'DeafOrHardOfHearing', 'BlindOrVisionDifficulty', 'DifficultyConcentrating',
    'DifficultyWalking', 'DifficultyDressingBathing', 'DifficultyErrands',
    'SmokerStatus', 'ECigaretteUsage', 'ChestScan', 'RaceEthnicityCategory',
    'AlcoholDrinkers', 'HIVTesting', 'FluVaxLast12', 'PneumoVaxEver',
    'TetanusLast10Tdap', 'HighRiskLastYear', 'CovidPos'
]

le = LabelEncoder()
for col in nominal_cols:
    df_numeric[col] = le.fit_transform(df_numeric[col].astype(str))
```

```python
# 누락된 값 채우기
# 숫자 열: 중앙값으로 대입하기
numeric_cols = ['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours', 'BMI', 'HeightInMeters', 'WeightInKilograms']
df_filtered[numeric_cols] = df_filtered[numeric_cols].fillna(df_filtered[numeric_cols].median())

# 서수 열: 최빈값으로 대입하기
ordinal_cols = ['GeneralHealth', 'AgeCategory']
df_filtered[ordinal_cols] = df_filtered[ordinal_cols].fillna(df_filtered[ordinal_cols].mode().iloc[0])

# 범주형 열: 최빈값으로 대입
categorical_cols = [
    'DifficultyConcentrating', 'DifficultyWalking',
    'LastCheckupTime', 'PhysicalActivities', 'HadAngina', 'HadStroke',
    'HadAsthma', 'HadSkinCancer', 'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease',
    'HadArthritis', 'HadDiabetes', 'DeafOrHardOfHearing', 'SmokerStatus', 'ChestScan',
    'AlcoholDrinkers'
]
df_filtered[categorical_cols] = df_filtered[categorical_cols].fillna(df_filtered[categorical_cols].mode().iloc[0])

print("대입 후 누락된 값\n", df_filtered.isna().sum())
print("최종 shape", df_filtered.shape)
```

```python
outlier_columns = []

for column in df_filtered.select_dtypes(include=['number']).columns: 
    Q1 = df_filtered[column].quantile(0.25)
    Q3 = df_filtered[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df_filtered[(df_filtered[column] < lower_bound) | (df_filtered[column] > upper_bound)]
    
    if not outliers.empty:
        outlier_columns.append(column)
        print(f"열 '{column}' 이상치 {len(outliers)}개 있음")

print("이상치 열", outlier_columns)
```

```python
from scipy import stats

outlier_cols_zscore = []

for column in df_filtered.select_dtypes(include=['number']).columns:
    z_scores = stats.zscore(df_filtered[column].dropna())
    outliers = (abs(z_scores) > 3).sum()

    if outliers > 0:
        outlier_cols_zscore.append(column)
        print(f"{column} 이상치 {outliers}개 있음")

print("Z-Score 이상치 열", outlier_cols_zscore)
```

```python
cap_physical = df_filtered['PhysicalHealthDays'].quantile(0.95)
cap_mental = df_filtered['MentalHealthDays'].quantile(0.95)
df_filtered['PhysicalHealthDays'] = df_filtered['PhysicalHealthDays'].clip(upper=cap_physical)
df_filtered['MentalHealthDays'] = df_filtered['MentalHealthDays'].clip(upper=cap_mental)

cap_height = df_filtered['HeightInMeters'].quantile(0.95)
cap_weight = df_filtered['WeightInKilograms'].quantile(0.95)
df_filtered['HeightInMeters'] = df_filtered['HeightInMeters'].clip(upper=cap_height)
df_filtered['WeightInKilograms'] = df_filtered['WeightInKilograms'].clip(upper=cap_weight)


Q1_sleep = df_filtered['SleepHours'].quantile(0.25)
Q3_sleep = df_filtered['SleepHours'].quantile(0.75)
IQR_sleep = Q3_sleep - Q1_sleep
lower_sleep = Q1_sleep - 1.5 * IQR_sleep
upper_sleep = Q3_sleep + 1.5 * IQR_sleep

Q1_bmi = df_filtered['BMI'].quantile(0.25)
Q3_bmi = df_filtered['BMI'].quantile(0.75)
IQR_bmi = Q3_bmi - Q1_bmi
lower_bmi = Q1_bmi - 1.5 * IQR_bmi
upper_bmi = Q3_bmi + 1.5 * IQR_bmi

df_filtered = df_filtered[
    (df_filtered['SleepHours'].between(lower_sleep, upper_sleep)) &
    (df_filtered['BMI'].between(lower_bmi, upper_bmi))
]

print("이상치 처리 후 최종 shape", df_filtered.shape)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report, confusion_matrix

def evaluate_model(model, model_name, X_train, y_train, X_test, y_test):
    # 모델 학습
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    y_train_prob = model.predict_proba(X_train)[:, 1]
    y_test_prob = model.predict_proba(X_test)[:, 1]

    train_acc = accuracy_score(y_train, y_train_pred)
    train_prec = precision_score(y_train, y_train_pred)
    train_rec = recall_score(y_train, y_train_pred)
    train_f1 = f1_score(y_train, y_train_pred)
    train_roc_auc = roc_auc_score(y_train, y_train_prob)

    test_acc = accuracy_score(y_test, y_test_pred)
    test_prec = precision_score(y_test, y_test_pred)
    test_rec = recall_score(y_test, y_test_pred)
    test_f1 = f1_score(y_test, y_test_pred)
    test_roc_auc = roc_auc_score(y_test, y_test_prob)

    # Print classification report
    print(f"{model_name} - 분류:\n", classification_report(y_test, y_test_pred))

    # Visualize confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(confusion_matrix(y_test, y_test_pred), annot=True, fmt="d", cmap="Blues")
    plt.xlabel("예측")
    plt.ylabel("실제 데이터")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()

    # Print results
    print(f"{model_name} 결과")
    print(f"Train Accuracy: {train_acc:.4f}")
    print(f"Train Precision: {train_prec:.4f}")
    print(f"Train Recall: {train_rec:.4f}")
    print(f"Train F1-Score: {train_f1:.4f}")
    print(f"Train ROC-AUC: {train_roc_auc:.4f}")
    print("-------------------------------")
    print(f"Test Accuracy: {test_acc:.4f}")
    print(f"Test Precision: {test_prec:.4f}")
    print(f"Test Recall: {test_rec:.4f}")
    print(f"Test F1-Score: {test_f1:.4f}")
    print(f"Test ROC-AUC: {test_roc_auc:.4f}")
    print("\n")
```

```python
df_filtered_numeric = df_filtered.copy()

object_cols = df_filtered_numeric.select_dtypes(include=['object']).columns.tolist()
print("Object columns:", object_cols)

binary_cols = [
    'Sex', 'PhysicalActivities', 'HadAngina', 'HadStroke',
    'HadAsthma', 'HadSkinCancer', 'HadCOPD', 'HadDepressiveDisorder',
    'HadKidneyDisease', 'HadArthritis', 'HadDiabetes', 'DeafOrHardOfHearing',
    'DifficultyConcentrating', 'DifficultyWalking',
    'ChestScan', 'AlcoholDrinkers'
]

for col in binary_cols:
    if col == 'Sex':
        df_filtered_numeric[col] = df_filtered_numeric[col].map({'Female': 0, 'Male': 1})
    elif col == 'HadDiabetes':
        df_filtered_numeric[col] = df_filtered_numeric[col].map({
            'No': 0,
            'Yes': 1,
            'No, pre-diabetes or borderline diabetes': 2,
            'Yes, but only during pregnancy (female)': 3
        })
    else:
        df_filtered_numeric[col] = df_filtered_numeric[col].map({'Yes': 1, 'No': 0})

# GeneralHealth
general_health_mapping = {
    'Poor': 0, 'Fair': 1, 'Good': 2, 'Very good': 3, 'Excellent': 4
}
df_filtered_numeric['GeneralHealth'] = df_filtered_numeric['GeneralHealth'].map(general_health_mapping)

last_checkup_mapping = {
    'Within past year (anytime less than 12 months ago)': 0,
    'Within past 2 years (1 year but less than 2 years ago)': 1,
    'Within past 5 years (2 years but less than 5 years ago)': 2,
    '5 or more years ago': 3
}
df_filtered_numeric['LastCheckupTime'] = df_filtered_numeric['LastCheckupTime'].map(last_checkup_mapping)

smoker_status_mapping = {
    'Never smoked': 0, 'Former smoker': 1, 'Current smoker - now smokes some days': 2,
    'Current smoker - now smokes every day': 3
}
df_filtered_numeric['SmokerStatus'] = df_filtered_numeric['SmokerStatus'].map(smoker_status_mapping)

age_category_mapping = {
    'Age 18 to 24': 0, 'Age 25 to 29': 1, 'Age 30 to 34': 2, 'Age 35 to 39': 3, 'Age 40 to 44': 4,
    'Age 45 to 49': 5, 'Age 50 to 54': 6, 'Age 55 to 59': 7, 'Age 60 to 64': 8, 'Age 65 to 69': 9,
    'Age 70 to 74': 10, 'Age 75 to 79': 11, 'Age 80 or older': 12
}
df_filtered_numeric['AgeCategory'] = df_filtered_numeric['AgeCategory'].map(age_category_mapping)

print("\n인코딩 후 결측값\n", df_filtered_numeric.isna().sum())

print(df_filtered_numeric.info())
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df_filtered_numeric.drop(columns=["HadHeartAttack"])
y = df_filtered_numeric["HadHeartAttack"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42, stratify=y)

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("X_train shape:", X_train_res.shape)
print("y_train shape:", y_train_res.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
print("-------------------------------")
print("클래스 분포 y_train_res:", y_train_res.value_counts())
print("-------------------------------")
print("클래스 분포 y_test:", y_test.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200, max_depth=10, min_samples_split=5, min_samples_leaf=2, random_state=42, class_weight='balanced'
)
evaluate_model(rf_model, "Random Forest", X_train_res, y_train_res, X_test, y_test)

rf_model.exp
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(
    max_depth=10, min_samples_split=5, min_samples_leaf=2, random_state=42, class_weight='balanced'
)
evaluate_model(dt_model, "Decision Tree", X_train_res, y_train_res, X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')

evaluate_model(log_model, "Logistic Regression", X_train_res, y_train_res, X_test, y_test)
```
